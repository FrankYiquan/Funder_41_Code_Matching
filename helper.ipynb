```python
import re

def get_country_city(funder_name):
    # Find the last (...) group
    match = re.findall(r"\(([^()]*)\)", funder_name)
    if match:
        parts = match[-1].split(",")
        if len(parts) >= 2:
            country = parts[0].strip()
            city = ",".join(parts[1:]).strip()
            return country, city
    return "not_found", "not_found"


print(get_country_city("Alphamade (Berlin) (Germany, Berlin)"))
```
